```python
import pandas as pd
import numpy as np
import re
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
```

```python
# read data
data = pd.read_csv("../data/raw/steam_reviews-ENGLISH.csv")
```

```python
data.head()
```

```python
# function developed by RafaÅ‚ WÃ³jcik from https://towardsdatascience.com/unsupervised-sentiment-analysis-a38bf1906483
# with some minor changes

def clean_text(text):
    ''' Pre process and convert texts to a list of words 
    method inspired by method from eliorc github repo: https://github.com/eliorc/Medium/blob/master/MaLSTM.ipynb'''
    
    text = str(text)
    text = text.lower()

    # Clean the text
    text = re.sub(r"[^A-Za-z0-9^,!?.\/'+]", " ", text)
    text = re.sub(r"\+", " ", text)
    text = re.sub(r",", " ", text)
    text = re.sub(r"\.", " ", text)
    text = re.sub(r"!", " ", text)
    text = re.sub(r"\?", " ", text)
    text = re.sub(r":", " : ", text)
    text = re.sub(r"\s{2,}", " ", text)
    text = re.sub(r"[0-9]", " ", text)

    text = text.split()

    return text
```

```python
data['review'] = data['review'].apply(lambda x: clean_text(x))
```

```python
# dealing with words with too many characters
data['review'] = data['review'].apply(lambda x: [i if len(i)<27 else np.nan  for i in x])
```

```python
data['review'] = data['review'].apply(lambda x: ' '.join(x))
```
